```python
Significantcoldf0=['Brand','Model','Accel(sec)','TopSpeed(km/h)','Range(km)','Efficiency(Wh/km)','FastCharge(km/h)',
                   'RapidCharge','PowerTrain','PlugType','BodyStyle','Segment','Seats','PriceEuro']
categoricalcoldf0=['Brand','RapidCharge','PowerTrain','PlugType','BodyStyle','Segment','Seats']
Numericalcoldf0=['Accel(sec)','TopSpeed(km/h)','Range(km)','Efficiency(Wh/km)','FastCharge(km/h)','Seats','PriceEuro']

Significantcoldf1=['County','Model Year','Make','Model','Electric Vehicle Type',
                   'Clean Alternative Fuel Vehicle (CAFV) Eligibility','Electric Range','Base MSRP']
categoricalcoldf1=['Model Year','Make','Electric Vehicle Type','Clean Alternative Fuel Vehicle (CAFV) Eligibility','County']
Numericalcoldf1=['Model Year','Electric Range','Base MSRP']
```

Cleaning

```python
#finding null and zero values in both dataframe
for g in df0.columns:
            colName = g
            print("\n***"+str(g)+"***")
            print("Null Values: ", df0[colName].isnull().sum())
            print("Zero Values: ", (df0[colName]==0).sum())

for g in df1.columns:
            colName = g
            print("\n***"+str(g)+"***")
            print("Null Values: ", df1[colName].isnull().sum())
            print("Zero Values: ", (df1[colName]==0).sum())
```

```python
#Replacing null and zero values of numerical column in both dataframe
for g in Numericalcoldf0:
            print("\n***"+str(g)+"***")
            print("*Original Count*")
            print("Null Values: ",df0[g].isnull().sum())
            print("Zero Values: ", (df0[g]==0).sum())
            df0[g] = df0[g].fillna(df0[g].median())
            df0[g] = np.where(df0[g]==0, df0[g].median(), df0[g])
            print("*Cleaned Count*")
            print("Null Values: ",df0[g].isnull().sum())
            print("Zero Values: ", (df0[g]==0).sum())  

for g in Numericalcoldf1:
            print("\n***"+str(g)+"***")
            print("*Original Count*")
            print("Null Values: ",df1[g].isnull().sum())
            print("Zero Values: ", (df1[g]==0).sum())
            df1[g] = df1[g].fillna(df1[g].median())
            df1[g] = np.where(df1[g]==0, df1[g].median(), df1[g])
            print("*Cleaned Count*")
            print("Null Values: ",df1[g].isnull().sum())
            print("Zero Values: ", (df1[g]==0).sum())
```

```python
#Replacing null value of column County
g='County'
print("*Original Count*")
print("***"+str(g)+"***")
print("Null Values: ", df1[g].isnull().sum())
print("Zero Values: ", (df1[g]==0).sum())
df1[g] = np.where(df1[g].isnull(),'Other', df1[g])
print("\n*Cleaned Count*")
print("***"+str(g)+"***")
print("Null Values: ", df1[g].isnull().sum())
print("Zero Values: ", (df1[g]==0).sum())            
```

```python
# Checking for Obvious error in categorical columns
for g in categoricalcoldf0:
  print("\n")
  print(df0.groupby([g])['Id'].count())
```

```python
for g in categoricalcoldf1:
  print("\n")
  print(df1.groupby([g])['VIN (1-10)'].count())
```

```python
colName='RapidCharge'
print("*Original Value*")
print(df0[colName].unique())
df0[colName] = df0[colName].map({"Rapid charging possible":'Yes', "Rapid charging not possible":'No'})
print("*Cleaned Count*")
print(df0[colName].unique())
```

```python
colName='Clean Alternative Fuel Vehicle (CAFV) Eligibility'
print("*Original Value*")
print(df1[colName].unique())
df1[colName] = df1[colName].map({"Clean Alternative Fuel Vehicle Eligible":'Yes', "Not eligible due to low battery range":'No'})
print("*Cleaned Count*")
print(df1[colName].unique())
```

```python
df0['Seats']=df0['Seats'].astype(int)
df1['Model Year']=df1['Model Year'].astype(int)
```

```python
# Correlation table and heatmap of Numerical columns 
dfe = df0[Numericalcoldf0]
print("\n*** Correlation Table ***")
pd.options.display.float_format = '{:,.3f}'.format
dfe=dfe.corr()
print(dfe)
print("\n*** HeatMap ***")       
plt.figure(figsize=(8,8))
ax = sns.heatmap(dfe, annot=True, cmap="PiYG")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom+0.5, top-0.5)
plt.show()
```
